```python
import pandas as pd

# File path (adjust as needed)
file_path = r"C:\Users\patil\Downloads\food recipe new\dataset(1).csv"

# Try reading the file with different encodings
try:
    recipe_df = pd.read_csv(file_path, encoding='ISO-8859-1')
except UnicodeDecodeError:
    recipe_df = pd.read_csv(file_path, encoding='cp1252')

# Convert the 'Ingredients' column from string to lists
# Assuming ingredients are stored as comma-separated strings
recipe_df['Ingredients'] = recipe_df['Ingredients'].apply(lambda x: [ingredient.strip().lower() for ingredient in x.split(',')])

# Function to recommend recipes based on input ingredients
def recommend_recipes(ingredients):
    # Convert input ingredients to lowercase for consistency
    ingredients = [ingredient.lower().strip() for ingredient in ingredients]

    # Function to check if all input ingredients are present in the recipe
    def has_all_ingredients(recipe_ingredients):
        # Check if all input ingredients are in the recipe
        return all(ingredient in recipe_ingredients for ingredient in ingredients) and len(recipe_ingredients) == len(ingredients)

    # Filter recipes that match the input ingredients
    recommendations = recipe_df[recipe_df['Ingredients'].apply(has_all_ingredients)]

    # Return only the matching recipes
    return recommendations[['Recipe Name', 'Recipe Information']]

# Example usage
if __name__ == "__main__":
    # Take input from the user as a comma-separated string of ingredients
    user_input = input("Enter ingredients (comma-separated): ")
    input_ingredients = [item.strip() for item in user_input.split(',')]

    # Get recipe recommendations
    recommendations = recommend_recipes(input_ingredients)

    # Print the recommended recipes
    if not recommendations.empty:
        print("Top Recipe Recommendations:")
        for index, row in recommendations.iterrows():
            print(f"Recipe Name: {row['Recipe Name']}")
            print(f"Recipe Information: {row['Recipe Information']}")
            print('-' * 30)
    else:
        print("No matching recipes found.")
```
